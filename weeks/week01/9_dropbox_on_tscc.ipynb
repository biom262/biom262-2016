# Get Dropbox on TSCC

The nice thing about the Jupyter notebooks is that you get to see all the input inline. But what about the `pdf` or `svg` or other high-resolution images that you want to use for your publications? Those don't show up in Jupyter and the images you copy/paste out of it look really bad when you put them up on a slide. So, what do you do? Well, you could `scp` (or `pscp` for you Windows folks) but you'll soon get tired of doing that for ***eeeeevvvvvvverrrrrrrrrryyyyyyyy*** figure.

Enter Dropbox. If you set up Dropbox on TSCC, you can save your figures into `~/Dropbox/` and then you'll have all your figures on your local computer, moments after you make them on TSCC! It's magic\*! And by "magic" we really mean "Software Engineering."

These instructions are taken from the [official Dropbox installation instructions](https://www.dropbox.com/install?os=lnx), but with outputs and pictures :)
